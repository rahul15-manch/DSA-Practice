```python
#fruits into basket
def fruits_into_basket(fruits):
    left=0
    basket={}
    max_len=0
    for right in range(len(fruits)):
        if fruits[right] in basket:
            basket[fruits[right]]+=1
        else:
            basket[fruits[right]]=1
        while len(basket)>2:
            basket[fruits[left]]-=1
            if basket[fruits[left]]==0:
                del basket[fruits[left]]
            left+=1
        max_len=max(max_len,right-left+1)
    return max_len

print(fruits_into_basket([0,1,2,2]))
```

```python
#permutation in string
def permutation(s1,s2):
    if len(s1)>len(s2):
        return False
    freq1={}
    freq2={}
    window=len(s1)
    for ch in s1:
        freq1[ch]=freq1.get(ch,0)+1
    for i in range(len(s2)):
        freq2[s2[i]]=freq2.get(s2[i],0)+1
        if i>=window:
            left_char=s2[i-window]
            if freq2[left_char]==1:
                del freq2[left_char]
            else:
                freq2[left_char]-=1
        if freq1==freq2:
            return True
    return False

print(permutation("ab","ebaidooo"))
```
